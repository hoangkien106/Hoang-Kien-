```python
#W5A1
#Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) ->int:
def Max_Of_Two(a:int, b:int) -> int:
   if a > b:
      return a
   return b
assert Max_Of_Two(12,34) == 34
assert Max_Of_Two(0,-12) == 0
assert Max_Of_Two(4,4) == 4
assert Max_Of_Two(-100,0) == 0 
assert Max_Of_Two(0,1) == 1
print("tất cả kết quả đúng")
```

```python
#W5A2 Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]:
def swap(a:int,b:int) ->(int,int):
    return b,a
assert swap(1,2) == (2, 1)
assert swap(0,10) == (10,0)
assert swap(-3, -33) == (-33,-3)
assert swap(1,10) == (10,1)
assert swap(0,0) == (0,0)
print(f"Tất cả đều đúng ")
```

```python
#W5A3 : Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 
def is_prime_number(n:int) -> bool :
    if n < 2 :
        return False 
    for i in range(2,int(n**0.5)+1):
        if n % i == 0 :
            return False
    return True
assert is_prime_number(2) == True
assert is_prime_number(0) == False
assert is_prime_number(17)== True
assert is_prime_number(-9) == False
assert is_prime_number(53) == True
print(f"Tất cả đúng")
```

```python
#W5A4 : VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 
def is_perfect_number(n:int) -> bool :
    if n <= 2 :
        return False 
    tong = 0 
    for i in range(1,n):
        if n % i == 0 :
            tong += i
    return tong == n 
assert is_perfect_number(6) == True
assert is_perfect_number(28)== True
assert is_perfect_number(496) == True
assert is_perfect_number(128)== False
print(f"Tất cả đúng")
```

```python
#W5A5:Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1
def find_first_index(nums:list(), k:int) -> int :
    for index,value in enumerate(nums):
        if value == k:
            return index
    return -1
assert find_first_index((1,2,3,4,3),3) == 2
assert find_first_index((9,12,45,66,12),12) == 1
assert find_first_index((325,24,252,15),1) == -1 
```

```python
#W5A6: Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)
def _giaithua(n:int) -> int :
    s = 1 
    for i in range(1,n+1):
        s *= i 
    return s
assert _giaithua(3) == 6
assert _giaithua(4) == 24
assert _giaithua(6) == 720
```
