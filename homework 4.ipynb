```python
#W4A17 Nhập vào số nguyên dương a, in ra bảng cửu chương của a 
a = int(input())
for i in range(10):
    s = a * i
    print(f"{a}*{i}={s}")
```

```python
#W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 
a, b = map(int,input().split())
s = min(a,b)
for i in range(1,s+1):
    if s % i == 0 :
        print(i)
```

```python
#W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không
n = int(input())
while n % 2 == 0:
    n //= 2 
if n == 1 :
    print(f"n có dạng 2^k")
else :
    print(f"n ko có dạng đó")
    
```

```python
#W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n 
n = int(input())
n = abs(n)
tong = 0
while n >0 :
    tong += n%10
    n //= 10
print(tong)
    
```

```python
#W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 
n = int(input())
chan = 0 
le = 0
while n > 0 :
    donvi = n % 10
    if donvi % 2 == 0 :
        chan += 1 
    else :
        le += 1 
    n //= 10
print(f"số n có {chan} số chẵn và {le} số lẻ")
```

```python
#W4A23
#  Nhập n
# Cho S(k) = 1 + 2 + 3 + … + k
# Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 
n = int(input())
s = 0 
k = 0
while s + (k+1) < n :
    k += 1
    s += k
print(f"Tổng s lớn nhất là :{s}")
print(f"Số k lớn nhất là : {k}")
```

```python
#W4A24
# Nhập vào A, tìm n nhỏ nhất sao cho
# 1 + 1/2 + 1/3 + ... + 1/n > A
A = int(input())
n = 0 
tong = 0 
while tong < A :
    n += 1
    tong += 1/n
print(f"số n nhỏ nhất là : {n}")
print(f"Tổng lớn nhất là : {tong}")
```
