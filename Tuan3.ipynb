```python
#W3A1
n = int(input("Nhập số nguyên :"))
if 0 <= n <= 9 :
    print(f"số ngược lại của n là :{n}")
elif 10 <= n <= 99 :
    don_vi = n % 10
    chuc = n // 10 
    a = don_vi * 10 + chuc
    print(f"số ngược lại của n là :{a}")
elif 100 <= n <= 999 :
    don_vii = n % 10
    hangchuc = ( n // 10) % 10
    hangtram = n // 100
    b = don_vii * 100 + hangchuc * 10 + hangtram
    print(f"số ngược lại của n là : {b}")
elif 1000 <= n <= 9999 :
    donn_vi = n % 10
    hangchucc = ( n // 10) % 10 
    hangtramm = (n // 100) % 10
    hangnghin = n // 1000
    c = donn_vi * 1000 + hangchucc * 100 + hangtramm * 10 + hangnghin
    print(f"số ngược lại của n là : {c}")
else : 
    print(f"chưa biết")
```
