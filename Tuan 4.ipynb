```python
#W4A1
n = int(input("Nhập số nguyên n"))
if n >= 1000 :
    print("nhập số nhỏ hơn 1000 !")
else :
    s = 0
    for i in range(1, n + 1):
        s += i
    print(f"Tổng các số nguyên từ 1 đến n là: {s}")
    
        
```

```python
#W4A2
import math
n = 0 
while n <= 0 :
    n = int(input("Nhập số nguyên dương :"))
so_nguyen_to = True
if n < 2 :
    so_nguyen_to = False
else : 
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0 :
            so_nguyen_to = False
        break
if so_nguyen_to :
    print(f"{n} là số nguyên tố")
else :
    print(f"{n} không là số nguyên tố")
      
```

```python
#W4A4
n = int(input())
so_duong = abs(n)
s = 0
if n == 0 :
    s = 1
else :
    while so_duong > 0 :
        so_duong //= 10
        s += 1
print(f"số chữ số của {n} là: {s}")
```

```python
#W4A6
a, b = map(int,input().split())
tong = 0 
for i in range(a,b+1):
    so_nguyen_to = True
    if i >= 2 :
        for j in range(2,int(i**0.5)+1):
            if i % j == 0 :
                so_nguyen_to = False
                break
        if so_nguyen_to :
            tong += i
print(f"Tổng các số nguyên tố từ {a} đến {b} là : {tong}")




    



```
